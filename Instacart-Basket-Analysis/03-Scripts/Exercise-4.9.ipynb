# 06. Creating Line Charts

```python
#Applying the “seed” to ensures generating the same set of random numbers but in reproducible pattern
np.random.seed(4)
```

```python
#Create a list named 'dev' to store boolean values to test np.random.rant() based on less than or equal to 0.7
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
#Crating a dataframe 'big' to store 70 percent of the 'ords_prods_merge' data that is 'True'
#Store 70% of the sample in the dataframe 'big'
big = ords_prods_merge[dev]
```

```python
#Crating a dataframe 'small' to store 70 percent of the 'ords_prods_merge' data that is 'False'
#Store 30% of the sample in the dataframe 'small'
small = ords_prods_merge[~dev]
```

```python
#Checking the length of the 'ords_prods_merge' dataframe
len(ords_prods_merge)
```

```python
#Checking the length of the 'big' & 'small' dataframes
len(big) + len(small)
```

```python
#Creating a subset data, a new dataframe, df_2, that contains only the “prices” and “orders_day_of_week” columns from your small subset.
df_2 = small[['orders_day_of_week','prices']]
```

```python
#Creating a Line Chart for the 'orders_day_of_week'& 'prices' columns and assigning it to Line object
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
#Exporting and saving the figure -the line object
line.figure.savefig(os.path.join(path, '04-Analysis','Visualizations', 'line_ord_day_week_prices2.png'))
```
