# 04. User-defined Functions

```python
#Create a subset of the data
df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
#Creating a function named "price_label"
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
price_label
```

```python
#Applying the function to the dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Checking the function application on the dataframe
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# 05. The loc() Functions

```python
#Creating conditions for for the loc() funtion
#Condition for High-range product
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Creating conditions for for the loc() funtion
#Condition for Mid-range product
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
#Creating conditions for for the loc() funtion
#Condition for Low-range product
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Checking the function application on the dataframe
df['price_range_loc'].value_counts(dropna = False)
```

# 07. If-Statements with For-Loops

```python
#printing the frequency of the “orders_day_of_week” - to find the busiest and slowest days of sales
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Using the for-loop to step through values in the "orders_day_of_week" column to categorize the values
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Checking the 'result' list created
result
```

```python
#Combining the 'result' with the 'ords_prods_merge' dataframe - creating a new column named 'busiest_day' in the 'ords_prods_merge' dataframe
ords_prods_merge['busiest_day'] = result
```

```python
#Checking the frequency of the 'orders_day_of_week' in the context of the 'busiest_day' -the new column
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# 08. Task 4.7

```python
#Create a new column named 'Busiest_days' that contains three different values: Busiest days, Slowest days, & Regular days
result_2 = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regular days")
```

```python
result_2
```

```python
#Combining the 'results_2' with the 'ords_prods_merge' dataframe, by
#Creating a new column whihtin the 'ords_prods_merge' dataframe named "Busiest_days"
ords_prods_merge['Busiest_days'] = result_2
```

```python
#Checking the frequency of the 'orders_day_of_week' in the context of the 'Busiest days' -the new column
ords_prods_merge['Busiest_days'].value_counts(dropna = False)
```
