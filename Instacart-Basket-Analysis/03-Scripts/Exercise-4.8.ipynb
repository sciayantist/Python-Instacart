# 06. Aggregating Data

```python
#Single Aggregation (mean)
#Aggregate the 'order_number' column in the 'df' dataframe grouped by 'department_id'
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#Multiple Aggregations (mean', 'min', 'max)
#Aggregate the 'order_number' column in the 'df' dataframe grouped by 'department_id' 
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# 07. Aggregating Data with transform()

```python
#Creating a new column named 'max_order' in the 'df' dataframe, Grouped by 'user_id' column &
#Applying the transform function on the 'order_number' column to Aggregate by Max. 
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

# 08. Deriving Columns with loc()
##### Creating falgs - Assinging loyalty status in the new column named 'loyalty_flag' based on the rows of the 'max_order' column in the 'ords_prods_merge' dataframe

```python
#Assinging loyalty status - (> 40)  then 'Loyal customer'
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Assinging loyalty status -  (<= 40 &  > 10) then 'Regular customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Assinging loyalty status - (<= 10) then 'New customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Checking the new column 
df['loyalty_flag'].value_counts(dropna = False)
```

```python
#Checking the a list of columns within a the dataframe
df[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#Apply the transform () funtion on the 'days_since_prior_order' column to generate the median 'days_since_prior_order' for each user
#Assign the results in a new column names 'median_days_since_last_order'
ords_prods_merge['median_days_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Deriving anew column named 'order_frquency_flag', with loc()
#Setting up criteria - 'Regular customer' <= 20 &  > 10
ords_prods_merge.loc[(ords_prods_merge['median_days_since_last_order'] <= 20) & (ords_prods_merge['median_days_since_last_order'] > 10), 'order_frquency_flag'] = 'Regular customer'
```

```python
#Deriving anew column named 'order_frquency_flag', with loc()
#Setting up criteria - 'Frequent customer' <= 10
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] <= 10, 'order_frquency_flag'] = 'Frequent customer'
```
