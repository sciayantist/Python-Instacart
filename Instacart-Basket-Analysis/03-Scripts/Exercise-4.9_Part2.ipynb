```python
#Applying the “seed” to ensures generating the same set of random numbers but in reproducible pattern
np.random.seed(4)
```

```python
#Create a list named 'dev' to store boolean values to test np.random.rant() based on less than or equal to 0.7
dev = np.random.rand(len(ords_prods_customers_merge)) <= 0.7
```

```python
dev
```

```python
#Crating a dataframe 'big' to store 70 percent of the 'ords_prods_customers_merge' data that is 'True'
#Store 70% of the sample in the dataframe 'big'
big = ords_prods_customers_merge[dev]
```

```python
#Crating a dataframe 'small' to store 70 percent of the 'ords_prods_customers_merge' data that is 'False'
#Store 30% of the sample in the dataframe 'small'
small = ords_prods_customers_merge[~dev]
```

```python
#Checking the length of the 'ords_prods_customers_merge' dataframe
len(ords_prods_customers_merge)
```

```python
#Checking the length of the 'big' & 'small' dataframes
len(big) + len(small)
```

```python
#Creating a subset data, a new dataframe, df_2, that contains only the “prices” and “order_hour_of_day” columns from the small subset.
df_2 = small[['order_hour_of_day','prices']]
```

```python
#Checking the dataframe
df_2.head()
```

```python
#Creating a Line Chart for the 'order_hour_of_day' & 'prices' columns and assigning it to line object
line_prices_hour = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

```python
#Exporting and saving the figure -the line object
line_prices_hour.figure.savefig(os.path.join(path, '04-Analysis','Visualizations', 'line_prices_hour.png'))  
```

# Observation 5.
#### Line Chart:
#### x = 'order_hour_of_day' & y = 'prices'
#### There is a difference in Pricing patterns based on the order hour of the day.
#### Prices of items placed between 2AM and 5AM are higher compared to prices at other times of the day.
#### The majority of Instacart sales occur at 10AM, this time of day typically reflects the lowest prices.
