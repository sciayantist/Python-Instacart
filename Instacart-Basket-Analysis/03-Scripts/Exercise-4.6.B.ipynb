### Observation 5.

##### After viewing the df_prods dataframe, I noticed an extra "Unnamed: 0" column that was not needed. 
##### Removed this column, thus bringing the total number of rows in the df_prods dataframe to 49672 rows × 5 columns. 


##### Searched for a common column to merge the df_prods and orders_products_combined dataframes.
##### Found that both dataframes had the column 'product_id'.


##### The df_prods dataframe contains 49672 rows and 5 columns.
##### The orders_products_combined dataframe contains 32434489 rows x 10 columns.


##### Deciding on how to combine the data
##### Concatenation is not a good choice because both data sets are not of the same length.

##### Full outer join might not be the most suitable approach because the data frames are of different sizes. 
##### The left or right join may be not the best option because one of the dataframes will be joined only if there is a matche. This approach may result in the loss of some data from one of the dataframes, while other data maybe missing. 


##### Inner join is the choice that make sense. 
##### That's because we need the product information mapped to the orders_products_combined dataframe. 
##### That way, it helps us to answer many business questions about customers and their shopping behaviors. 
##### This approach results in a merge where only the rows that have a match in both dataframes will be included in the final merged dataframe. All other rows will be discarded.

##### Create a new dataframe, ords_prods_merge, to merge the orders_products_combined dataframe and df_prods dataframe together, using inner join.
